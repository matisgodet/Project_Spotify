```python
def open_file(file):
    """Opens json file and extracts the data as embedded dictionaries/lists."""
    with open(file, "r", encoding="utf-8") as json_data:
        data = json.load(json_data)
    return data

def playlist_lister(data):
    """Takes the data and extracts the playlist names, listing them one by one."""
    for playlist in data:
        print(playlist["name"])
    return None

def playlist_isolater(data):
    data_filtered = [playlist for playlist in data if playlist["name"].startswith("Playlist no.")]
    return data_filtered

data_raw = open_file(json_file)["playlists"]
data = playlist_isolater(data_raw)
```

```python
def unique_tracks_lister(data_list):
    unique_tracks = []
    seen_songs = set()
    for playlist in data_list:
        for song in playlist["items"]:
            if song not in seen_songs:
            # print(song["track"])
                unique_tracks.append(song["track"])
            else:
                seen_songs.add(song)
    return unique_tracks

unique_tracks = unique_tracks_lister(data)
```

```python
def get_track_tags(track_name, artist_name):

    url = "http://ws.audioscrobbler.com/2.0/"

    params = {
        "method": "track.getTopTags",
        "track": track_name,
        "artist": artist_name,
        "api_key": API_KEY,
        "format": "json"
    }

    r = requests.get(url, params=params)
    data = r.json()
    # print(data)
    # print(artist_name)
    tags = data["toptags"]["tag"]
    # print("Song tags:")
    # print(tags)
    return [tag["name"] for tag in tags[:5]]

def get_artist_tags(artist_name):

    url = "http://ws.audioscrobbler.com/2.0/"

    params = {
        "method": "artist.getTopTags",
        "artist": artist_name,
        "api_key": API_KEY,
        "format": "json"
    }

    r = requests.get(url, params=params)
    data = r.json()
    # print(data)
    # print(artist_name)
    tags = data["toptags"]["tag"]
    # print(tags)
    return [tag["name"] for tag in tags[:5]]
```
